```python
import pandas as pd
from geopy.distance import geodesic

# Load the dataset
file_path = '/content/DJIFlightRecord_2020-05-09_(15-36-27).csv'
data = pd.read_csv(file_path)

# Assume data is in chronological order

# Calculate BatteryPercentageLanding - Get the last value in the dataset
battery_percentage_landing = data['BATTERY.chargeLevel'].iloc[-1]

# FlightDurationMin - Assuming total flight time is already in seconds, convert to minutes
flight_duration_min = data['OSD.flyTime [s]'].iloc[-1] / 60

# MaximumAltitudeFt - Directly use the maximum value recorded
maximum_altitude_ft = data['OSD.heightMax [ft]'].max()

# Calculate DistanceFromPilot - assuming home latitude and longitude are known
home_latitude, home_longitude = data['HOME.latitude'].iloc[0], data['HOME.longitude'].iloc[0]
data['distance_from_pilot_ft'] = data.apply(lambda row: geodesic((home_latitude, home_longitude), (row['OSD.latitude'], row['OSD.longitude'])).feet, axis=1)
distance_from_pilot_ft = data['distance_from_pilot_ft'].max()

# NumberOfFlightsSubmitted - This value needs to be maintained externally
number_of_flights_submitted = 1

# SafetyScore calculation
safety_score = (0.2 * (battery_percentage_landing / 100) +
                -0.1 * (flight_duration_min / 30) +
                -0.0001 * (distance_from_pilot_ft / 500) +
                -0.0001 * (maximum_altitude_ft / 400) +
                0.05 * (number_of_flights_submitted / 50))

print(f'Safety Score: {safety_score}')
```
